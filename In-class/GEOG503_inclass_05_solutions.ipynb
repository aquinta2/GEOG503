```python
import arcpy
arcpy.env.workspace = r"C:\geog503\inclass05"
arcpy.env.overwriteOutput = True
```

```python
fc_list = arcpy.ListFeatureClasses()
for fc in fc_list:
    desc = arcpy.Describe(fc)
    baseName = desc.baseName
    out_file = "Results/" + baseName + "_buffer.shp"
    shapeType = desc.shapeType
    if shapeType == "Point":
        arcpy.Buffer_analysis(fc,out_file,1000)
    elif shapeType == "Polyline":
        arcpy.Buffer_analysis(fc,out_file,2000)
    if shapeType == "Polygon":
        arcpy.Buffer_analysis(fc,out_file,3000)
```
