# Ploting raster data

```python
import arcpy
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Plotting without histogram streching
arcpy.env.workspace = r"C:\geog503\data"
image = arcpy.RasterToNumPyArray("dem.tif")
plt.imshow(image, interpolation=None)
plt.show()
```

```python
# Plotting with histogram streching
arcpy.env.workspace = r"C:\geog503\data"
image = arcpy.RasterToNumPyArray("dem.tif",nodata_to_value=0)
vmin = np.min(image[image > 0])
vmax = np.max(image)
print("min = {}, max = {}".format(vmin,vmax))
plt.imshow(image, interpolation=None,vmin=vmin, vmax=vmax)
plt.show()
```
